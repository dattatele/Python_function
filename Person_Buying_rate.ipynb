```python
def selection_person():
    data = pd.DataFrame(Booking_Event_2017_ScleanNeg)
    selected_name = [name for name in data.Full_Name.unique()]
    for s in selected_name:
        select_person = pd.DataFrame(data[data.Full_Name == s])
        if select_person.bookings_id.nunique() > 1:
            booking_id = [id  for id in select_person.bookings_id.unique()]
            Selected_booking = pd.DataFrame(select_person[select_person.bookings_id == booking_id])
            Show_Person = pd.DataFrame(Selected_booking.drop_duplicates(subset=['events_id'], keep=False))
            return show_Person
        return select_person
            
        
```

```python
def selection_name(name):
    data = pd.DataFrame(Booking_Event_2017_ScleanNeg)
    select_person = pd.DataFrame(data[data.Full_Name == name])
    booking_id = [id  for id in select_person.bookings_id.unique()]
    if select_person.bookings_id.nunique() > 1:
        for bs in booking_id:
            if select_person[select_person.bookings_id == bs].bookings_id.count() > 1:
                testing = pd.DataFrame(select_person[select_person.bookings_id == bs])
                Show_Person = pd.DataFrame(testing.drop_duplicates(subset=['events_id'], keep=False))
                return Show_Person
    return select_person
```

```python
def person_maximum():
    Name = []
    Event = []
    Event_DateTime = []
    Booking_DateTime = []
    Amount = []
    data = pd.DataFrame(Booking_Event_2017_ScleanNeg)
    selected_name = [name for name in data.Full_Name.unique()]
    for s in selected_name:
        select_person = pd.DataFrame(data[data.Full_Name == s])
        booking_id = [id  for id in select_person.bookings_id.unique()]
        if select_person.bookings_id.nunique() > 1:
            for bs in booking_id:
                if select_person[select_person.bookings_id == bs].bookings_id.count() > 1:
                    testing = pd.DataFrame(select_person[select_person.bookings_id == bs])
                    Show_Person = pd.DataFrame(testing.drop_duplicates(subset=['events_id'], keep=False))
                    Name.append(Show_Person.Full_Name.unique())
                    Event.append(Show_Person.name_short.unique())
                    Event_DateTime.append(Show_Person.events_datetimes.unique())
                    Booking_DateTime.append(Show_Person.booking_date.unique())
                    Amount.append(Show_Person.amount_local_y.unique())
    return pd.DataFrame(list(zip(Name, Event, Event_DateTime, Booking_DateTime, Amount)))
```

```python
def person_maxi():
    Name = []
    Event = []
    Event_DateTime = []
    Booking_DateTime = []
    Amount = []
    data = pd.DataFrame(Booking_Event_2017_ScleanNeg)
    selected_name = [name for name in data.Full_Name.unique()]
    for s in selected_name:
        select_person = pd.DataFrame(data[data.Full_Name == s])
        booking_id = [id  for id in select_person.bookings_id.unique()]
        if select_person.bookings_id.nunique() > 1:
            for bs in booking_id:
                if select_person[select_person.bookings_id == bs].bookings_id.count() > 1:
                    testing = pd.DataFrame(select_person[select_person.bookings_id == bs])
                    Show_Person = pd.DataFrame(testing.drop_duplicates(subset=['events_id'], keep=False))
                    Name.append(Show_Person.Full_Name.unique())
                    Event.append(Show_Person.name_short.unique()[0])
                    Event_DateTime.append(Show_Person.events_datetimes.unique()[0])
                    Booking_DateTime.append(Show_Person.booking_date.unique()[0])
                    Amount.append(Show_Person.amount_local_y.unique()[0])
    return pd.DataFrame(list(zip(Name, Event, Event_DateTime, Booking_DateTime, Amount)))
```
